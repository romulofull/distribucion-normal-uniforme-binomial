# Problemas de distribución de probabilidad

**## Distribución binomial**

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

**Instalaciones**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import uniform
from scipy.stats import binom
```

```python
#Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo, tome el valor de 2
def dbinom(k, n, p=0.2, loc=0):
    return binom.pmf(k, n, p, loc=0)

print(dbinom(2, 10, 0.2))
```

```python
#Probability that a Binomial(10, 0.2) will take a value less than 2
def pbinom(k, n, p=0.2):
    return binom.cdf(k, n, p, loc=0)

print(pbinom(2, 10, 0.2))
```

```python

#Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9
def qbinom(q, n, prob=0.2):
    return binom.ppf(q, n, prob, loc=0)

print(qbinom(0.9, 10, 0.2))
```

```python
np.random.seed(42)
#Generate 2000 random values from a Binomial(10, 0.2) distribution
def rbinom(n, p, size):
    return binom.rvs(n, prob, loc=0, size=2000)

print(rbinom(10, 0.2, 2000))
```

**Graficas**

```python
size = 10
prob = 0.2
data = rbinom(2000, 10, 0.2)

pmf = dbinom(data, size, prob)
cdf = pbinom(data, size, prob)
ppf = qbinom(data, size, prob)

fig, axis = plt.subplots(3, 1, figsize = (8, 10))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")
plt.tight_layout()
plt.show()
```

**## Distribución normal**


```python
#Calcula la estimacion puntual de la distribución normal.
def dnorm(x, loc=0, scale=1):
    return norm.pdf(x, loc=0, scale=1)

print(dnorm(2, 0, 1))
```

```python
 #Calcula el acumulado de la distribución normal.
 #Retorna P(X <= q)
def pnorm(x, loc=0, scale=1):
    return norm.cdf(x, loc=0, scale=1)
pnorm(2, loc=0, scale=1)
```

```python
#  Calcula el cuantil (valor de X) para una probabilidad acumulada
def qnorm(q, loc=0, scale=1):
    return norm.ppf(q, loc=0, scale=1)

print(qnorm(0.9, 0, 1))
```

```python
#Genera 2000 variables aleatorias a partir de la distribución normal.
np.random.seed(42)
#Generate 2000 random values from a Binomial(10, 0.2) distribution
def rnorm(loc= 0, scale=1, size=2000, prob=0.2):
    return norm.rvs(loc=0, scale=1, size=2000)

print(rnorm(0, 1, 2000, 0.2))
```

**Gráficas**

```python

loc = 0
scale = 1

data = norm.rvs(loc, scale, size = 2000)

pdf = norm.pdf(data, loc, scale)
cdf = norm.cdf(data, loc, scale)
ppf = norm.ppf(data, loc, scale)

#Graficas
fig, axis = plt.subplots(3, 1, figsize = (8, 10))

axis[0].plot(data, pdf, "r.")
axis[1].plot(data, cdf, "b.")
axis[2].plot(data, ppf, "g.")

plt.tight_layout()
plt.show()
```

**Distribucion Uniforme.**

```python
#Calcula la densidad puntual de la distribución uniforme. probabilidad en x = 0.8
def dunif(x, loc=0, scale=1):
    return uniform.pdf(x, loc=0, scale=1)

dunif(0.8, loc=0, scale=1)
```

```python
#Calcula el valor acumulado de la distribución uniforme. Retorna P(X <= q). probabilidad acumulada hasta x = 0.8
def punif(x, loc=0, scale=1):
    return uniform.cdf(x, loc=0, scale=1)

punif(0.8, loc=0, scale=1)
```

```python
# Calcula el cuantil (valor de X) para una probabilidad acumulada dada en una distribución uniforme continua. valor de X tal que P(X <= x) = 0.9
def qunif(q, loc = 0, scale = 1):
    return uniform.ppf(q, loc=0, scale=1)

qunif(0.9, loc = 0, scale = 1)
```

```python
#generar 2000 valores aleatorios de una Uniforme
np.random.seed(42)
def qunif(loc=0, scale=1, size=2000):
    return uniform.rvs(loc=0, scale=1, size=200)

qunif(0, 1, 2000)
```

```python
#Graficas


#Generamos 2000 valores aleatorios
data = uniform.rvs(0, 1, size = 2000)

#Calculamos PDF, CDF y PPF
pdf = uniform.pdf(data, 0, 1)
cdf = uniform.cdf(data, 0, 1)
ppf = uniform.ppf(cdf, 0, 1)
fig, axis = plt.subplots(3, 1, figsize = (8, 10))

axis[0].plot(data, pdf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(cdf, ppf, "g-")

plt.tight_layout()
plt.show()

```
